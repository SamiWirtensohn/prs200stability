```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('rotation_ol.txt', delimiter='\t')

x = data[:,0]
y1 = data[:,2]
y2 = data[:,4]
z = y1 - y2
```

```python
def func1(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func1, x, y1)
popt

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
def func2(x, m, t):
    return m * x + t

popt2, pcov2 = curve_fit(func2, x, y2)
popt2
```

```python
plt.plot(x, y1, label= 'Autokollimator')
plt.plot(x, y2, label= 'Pollux')
plt.plot(x, z, label= 'Differenz')
#plt.plot(x, func1(x, *popt), 'r-', label='fit_Autokoll')
#plt.plot(x, func2(x, *popt2), 'g-', label='fit2')
plt.title('Rotation im Open Loop')
plt.xlabel('Messung')
plt.ylabel('Bogensekunden')
plt.legend()
#plt.savefig('0cl.png', dpi = 300)
plt.show()
```

```python
def func3(x, m, t):
    return m * x + t

popt3, pcov = curve_fit(func3, x, z)
popt3
```

```python
plt.plot(x, z, label= 'Differenz')
plt.plot(x, func3(x, *popt3), 'r-')
```
