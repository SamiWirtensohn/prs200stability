```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
test = np.loadtxt('test.txt', delimiter=' ')

x = test[:,0]
y = test[:,1]
```

```python
def func(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func, x, y)
popt

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
plt.plot(x, y, label= 'Autokollimator')
plt.plot(x, func(x, *popt), 'r-', label='fit')
plt.title('3 Nm Closed Loop')
plt.xlabel('Messung')
plt.ylabel('Bogensekunden')
plt.legend()
plt.show()
```
