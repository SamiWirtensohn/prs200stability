```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('3ol.txt', delimiter='\t')

x = data[:,0]
y1 = data[:,2]
y2 = data[:,4]
```

```python
def func(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func, x, y1)
popt

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
def func(x, m, t):
    return m * x + t

popt2, pcov2 = curve_fit(func, x, y2)
popt2
```

```python
plt.plot(x, y1, '.', label= 'Autokollimator')
plt.plot(x, y2, '.', label= 'Pollux')
plt.plot(x, func(x, *popt), 'r-', label='fit_Autokoll')
plt.plot(x, func(x, *popt2), 'g-', label='fit_Pollux')
plt.title('3 Nm Open Loop')
plt.xlabel('Messung')
plt.ylabel('Bogensekunden')
plt.legend()
#plt.savefig('3ol.png', dpi = 300)
plt.show()
```
