```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('3cl.txt', delimiter='\t')

x = data[:,0]
y1 = data[:,2]
y2 = data[:,4]
```

```python
def func1(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func1, x, y1)
popt

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
def func2(x, m, t):
    return m * x + t

popt2, pcov2 = curve_fit(func2, x, y2)
popt2
```

```python
plt.plot(x, y1, '.', label= 'Autokollimator')
plt.plot(x, y2, '.', label= 'Pollux')
plt.plot(x, func1(x, *popt), 'r-', label='fit_Autokoll')
plt.plot(x, func2(x, *popt2), 'g-', label='fit2')
plt.title('3 Nm Closed Loop')
plt.xlabel('Messung')
plt.ylabel('Bogensekunden')
plt.legend()
#plt.savefig('3cl.png', dpi = 300)
#plt.plot(x, func1(x, M, 0), 'k')
plt.show()
```
