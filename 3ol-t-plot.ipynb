```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('3ol-t.txt', delimiter='\t')

x = data[:,5]
y1 = data[:,2]
y2 = data[:,4]
```

```python
def func(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func, x, y1)
mA = popt[0]
tA = popt[1]
popt, mA, tA

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
popt2, pcov2 = curve_fit(func, x, y2)
mP = popt2[0]
tP = popt2[0]
popt2, mP, tP
```

```python
data = np.loadtxt('unabh.txt', delimiter='\t')

xt = data[:,0]
zt = data[:,2]
yt = data[:,4]
```

```python
poptU, pcovU = curve_fit(func, xt, zt)
mT = poptU[0]
tT = poptU[1]
poptU, mT, tT
```
