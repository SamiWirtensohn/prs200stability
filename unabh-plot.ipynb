```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('unabh.txt', delimiter='\t')

x = data[:,0]
z = data[:,2]
y = data[:,4]
```

```python
def func(x, m, t):
    return m * x + t

popt, pcov = curve_fit(func, x, z)
popt

#plt.plot(x, func(x, *popt), 'r-', label='fit')
```

```python
plt.plot(x, z, '.', label= 'Autokollimator')
plt.plot(x, func(x, *popt), 'r-', label='fit_Autokoll')
plt.title('Von Stage unabh√§ngiger Spiegel')
plt.xlabel('Zeit[min]')
plt.ylabel('Bogensekunden')
plt.legend()
plt.savefig('unabh.png', dpi = 300)
plt.show()
```
